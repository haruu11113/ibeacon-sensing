```python
import math
#各座標の入力
A = (0, 0, 1)
B = (0, 2, 1)
C = (2, 0, 1)
R = (math.sqrt(10), math.sqrt(10), math.sqrt(10))
z = 1.0
#分解
a1, a2, a3 = A
b1, b2, b3 = B
c1, c2, c3 = C
ra, rb, rc = R
```

```python
#AとBから式を作る　ー　①
def calc_AB():
    global a1, a2, b1, b2, a3, b3
    ABx = 2 * a1 - 2 * b1
    ABy = 2 * a2 - 2 * b2
    ABr = (rb ** 2 - ra ** 2) + (a1 ** 2 + a2 ** 2 + a3 ** 2) - (b1 ** 2 + b2 ** 2 + b3 ** 2) - ((2 * a3 - 2 * b3) * z)
    return ABx, ABy, ABr

#AとCから式を作る　ー　②
def calc_AC():
    global a1, a2, b1, b2, a3, b3
    ACx = 2 * a1 - 2 * c1
    ACy = 2 * a2 - 2 * c2
    ACr = (rc ** 2 - ra ** 2) + (a1 ** 2 + a2 ** 2 + a3 ** 2) - (c1 ** 2 + c2 ** 2 + c3 ** 2) - ((2 * a3 - 2 * c3) * z)
    return ACx, ACy, ACr

#BとCから式を作る　ー　③
def calc_BC():
    global a1, a2, b1, b2, a3, b3
    BCx = 2 * b1 - 2 * c1
    BCy = 2 * b2 - 2 * c2
    BCr = (rc ** 2 - rb ** 2) + (b1 ** 2 + b2 ** 2 + b3 ** 2) - (c1 ** 2 + c2 ** 2 + c3 ** 2) - ((2 * b3 - 2 * c3) * z)
    return BCx, BCy, BCr
```

```python
ABx, ABy, ABr = calc_AB()
ACx, ACy, ACr = calc_AC()
BCx, BCy, BCr = calc_BC()

#① - ② (yを削除)
ABx *= ACy
ABr *= ACy
ACx *= ABy
ACr *= ABy
#xを求める
x = ABx - ACx
r = ABr - ACr
x = r / x
#yを求める
ABx, ABy, ABr = calc_AB()
ABx *= x
ABr -= ABx
y = ABr / ABy
print(x, y) #0.5000000000000001 0.6
```
